```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add node in the beginning
    def AddBeginning(self, newdata):
        NewNode = Node(newdata)
        NewNode.nextval = self.headval
        self.headval = NewNode

    # Function to remove node by key
    def RemoveNode(self, Removekey):
        HeadVal = self.headval

        if HeadVal is not None:
            if HeadVal.dataval == Removekey:
                self.headval = HeadVal.nextval
                HeadVal = None
                return

        while HeadVal is not None:
            if HeadVal.dataval == Removekey:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval

        if HeadVal is None:
            return

        prev.nextval = HeadVal.nextval
        HeadVal = None


# create linked list named L1
L1 = LinkedList()

# add node
L1.AddBeginning("April")
L1.AddBeginning("Maret")
L1.AddBeginning("Februari")

print("Sebelum dilakukan remove")
L1.listprint()

print("-----------------------------")

# remove node by key "Maret"
L1.RemoveNode("Maret")

print("Setelah dilakukan remove dengan kunci 'Maret'")
L1.listprint()
```

```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class LinkedList:
    def __init__(self):
        self.headval = None

    # Print the linked list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # Function to add new node in the end
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode

    # Function to remove last node
    def RemoveEnd(self):
        last = self.headval

        if last is None:
            return

        if last.nextval is None:
            self.headval = None
            return

        while last.nextval is not None:
            if last.nextval.nextval is None:
                break
            prev = last
            last = last.nextval

        last.nextval = None


# create linked list named L1
L1 = LinkedList()

# add node
L1.AddInEnd("Januari")
L1.AddInEnd("Februari")
L1.AddInEnd("Maret")

print("Sebelum dilakukan remove")
L1.listprint()

print("----------------------------------")

# remove last node
L1.RemoveEnd()

print("Setelah dilakukan remove node terakhir")
L1.listprint()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Print list
    def display(self):
        temp = self.head
        while temp:
            print(temp.data)
            temp = temp.next

    # Add at end
    def add_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

    # Insert after specific key
    def insert_after(self, key, data):
        temp = self.head
        while temp:
            if temp.data == key:
                new_node = Node(data)
                new_node.next = temp.next
                new_node.prev = temp

                if temp.next:
                    temp.next.prev = new_node

                temp.next = new_node
                return
            temp = temp.next
        print("Key tidak ditemukan")

    # Remove by key
    def remove(self, key):
        temp = self.head

        while temp:
            if temp.data == key:
                # Jika node pertama
                if temp.prev is None:
                    self.head = temp.next
                    if self.head:
                        self.head.prev = None
                else:
                    temp.prev.next = temp.next
                    if temp.next:
                        temp.next.prev = temp.prev
                return
            temp = temp.next
        print("Key tidak ditemukan")


# ===== TEST =====
dll = DoublyLinkedList()

dll.add_end(1)
dll.add_end(2)
dll.add_end(3)

print("Sebelum penyisipan:")
dll.display()

dll.insert_after(2, 99)

print("\nSetelah sisip tengah:")
dll.display()

dll.remove(1)

print("\nSetelah hapus key 1:")
dll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Print list
    def display(self):
        if self.head is None:
            return

        temp = self.head
        while True:
            print(temp.data)
            temp = temp.next
            if temp == self.head:
                break

    # Add at end
    def add_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    # Remove last node
    def remove_end(self):
        if self.head is None:
            return

        # Jika hanya satu node
        if self.head.next == self.head:
            self.head = None
            return

        temp = self.head
        while temp.next.next != self.head:
            temp = temp.next

        temp.next = self.head


# ===== TEST =====
cll = CircularLinkedList()

cll.add_end(10)
cll.add_end(20)
cll.add_end(30)

print("Sebelum tambah akhir:")
cll.display()

cll.add_end(40)

print("\nSetelah tambah akhir:")
cll.display()

cll.remove_end()

print("\nSetelah hapus akhir:")
cll.display()
```
